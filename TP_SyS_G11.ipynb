## Simulaciones de señales

En los siguientes códigos se gráfican una señal senoidal y una señal cosenoidal superpuestas. Estas comparten periodo y frecuencia de muestreo, pero cada una tiene su frecuencia individual. Primero se osberva una continua y luego una discreta.

```python
T = 1 #Periodo
fs = 1000 #Frecuencia de muestreo
Frecuencia_seno = 2
Frecuencia_coseno = 2
t = np.linspace(0,T,T*fs, endpoint = False) #Eje del tiempo
x1 = np.sin(t*2*np.pi * Frecuencia_seno) #Señal seno
x2 = np.cos(t*2*np.pi * Frecuencia_coseno) #Señal coseno
plt.plot(t,x1,color="red")
plt.plot(t,x2, color = "blue")
plt.xlabel("Tiempo [s]")
plt.ylabel("x(t)")
plt.title("Señales continuas")
plt.grid()
plt.legend(["Señal senoidal", "Señal cosenoidal"], loc = "upper right")
plt.show()
```

```python
T = 1
fs = 80
Frecuencia_seno = 2
Frecuencia_coseno = 2
t = np.linspace(0,T,T*fs, endpoint = False)
x1 = np.sin(t*2*np.pi * Frecuencia_seno)
x2 = np.cos(t*2*np.pi * Frecuencia_coseno)
plt.stem(t,x1,linefmt="red")
plt.stem(t,x2, linefmt = "blue")
plt.xlabel("Tiempo [s]")
plt.ylabel("x(t)")
plt.title("Señales discretas")
plt.grid()
plt.legend(["Señal senoidal", "Señal cosenoidal"], loc = "upper right")
plt.show()
```

En los siguientes códigos se gráfican señales de escalón. Esta señal se encuentra centrada en b, por lo que para todo t mayor que b la señal vale 1. Primero se osberva una continua y luego una discreta.

```python
T = 1
fs = 1000
b = 0
t = np.linspace(-abs(b)-1,T*abs(b)+1,T*fs, endpoint = False)
x1 = np.heaviside(t + b, 0.5)
plt.plot(t,x1,color="red")
plt.xlabel("Tiempo [s]")
plt.ylabel("x(t)")
plt.title("Señal continua - Escalón")
plt.grid()
plt.legend(["u(t+b)"], loc = "upper left")
plt.show()
```

A continuación se presentan los códigos que nos permiten graficar la función rampa, primero en tiempo continuo y luego en tiempo discreto. Se utiliza la función "ramp(x)" que lo que hace es tomar un valor x como entrada y devuelve el valor de x si x es mayor que 0, y devuelve 0 si x es menor o igual a 0.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función rampa
def ramp(x):
    return np.maximum(0, x)

# Generar datos para graficar
T = 1
fs = 1000
Frecuencia_rampa = 2
x = np.linspace(-1, T + 1, T*fs, endpoint = False)
y = ramp(x)

# Graficar la función rampa
plt.plot(x, y, color="purple", linewidth=3)
plt.title('Función Rampa')
plt.xlabel('Tiempo [s]')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función rampa
def ramp(x):
    return np.maximum(0, x)

# Generar datos discretos para graficar
x_discrete = np.arange(-10, 11)  # Valores de x de -10 a 10
y_discrete = ramp(x_discrete)     # Calcular los valores correspondientes de la función rampa

# Graficar la función rampa de forma discreta
plt.stem(x_discrete, y_discrete, use_line_collection=True)
plt.title('Función Rampa (Discreta)')
plt.xlabel('x')
plt.ylabel('ramp(x)')
plt.grid(True)
plt.show()
```

```python
T = 1
fs = 1000
b = -1
t = np.linspace(-T*abs(b) - 1,T*abs(b) + 1,T*fs, endpoint = False)
x1 = np.heaviside((t + 1/2) + b, 0.5)
x2 = np.heaviside((t - 1/2) + b, 0.5)
plt.plot(t,x1-x2,color="red")
plt.xlabel("Tiempo [s]")
plt.ylabel("x(t)")
plt.title("Señal pulso cuadrado")
plt.grid()
plt.legend(["u(t + 0.5) - u(t - 0.5)"], loc = "upper left")
plt.show()
```

Para el siguiente código se tomaron los datos de un ECG a partir de una base de datos y se realizó el gráfico según la derivación que el usuario desee visualizar.

```python
directory = "/content/drive/MyDrive/SyS/"
ECGs = []
for ecgfilename in sorted(os.listdir(directory )):
    if ecgfilename.endswith(".dat"):
        ecg = wfdb.rdsamp(directory  + ecgfilename.split(".")[0])
        ECGs.append(ecg)
ECGs = np.asarray(ECGs, dtype="object")
def extraer_datos(dato):
  lista = ECGs[1][1]['sig_name']
  index = lista.index(dato)
  for i in ECGs[1][1]['sig_name']:
    if i == dato:
      Señal = []
      i=0
      while i<5000: #para cada fila de las 5000
        Señal.append(ECGs[0][0][i][index]) # el segundo 0 es porque quiero la derivacion I (hay información sobre las 12 derivaciones)
        i+=1
  return Señal
def visualizar(Signal, dato):
  yaxis = Signal
  T = 10
  fs = 500
  t = np.linspace(0,T,T*fs, endpoint = False)
  plt.figure(figsize=(25, 5))
  plt.plot(t,yaxis,color="red")
  plt.xlabel("Tiempo [s]")
  plt.ylabel("V [mV])")
  plt.title("Norwegian Endurance Athlete ECG")
  plt.legend([str("Derivacion " + dato)])
  plt.xticks(np.arange(0, T + 0.25, 0.25))
  plt.yticks(np.arange(-0.6, 0.7, 0.1))
  plt.grid(linestyle = '--', linewidth = 0.5)
  plt.show
print(ECGs[1][1]['sig_name']) #Estos son los tipos de mediciones #Para esto imprimi ECG, tenia que ver los elementos para poner los indices. Es un diccionario, adentro de una lista, adentro de una tupla
tipo_dato = input("Escriba que tipo de dato quiere visualizar: ")
visualizar(extraer_datos(tipo_dato), tipo_dato)
```
